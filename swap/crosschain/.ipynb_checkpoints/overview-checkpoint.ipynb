```python
import pool
import importlib
from copy import deepcopy
import matplotlib.pyplot as plt
importlib.reload(pool)

weights = [0.5, 0.5]
state = pool.create(weights=weights, max_allocation=0.8, weight_amplitude=0.65, slippage_rate=10, fee_max=10000000, fee_base=0)
pool.proportional_deposit(200, state)

%matplotlib widget
deposits = list(range(1, 57, 1))

def assets_rate(amounts: [int], state: pool.PoolState):
    pool.deposit(amounts, state)
    return state.assets_balances[0] / state.total_supply, state.assets_balances[1] / state.total_supply

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(deposits, 
                list(map(
                    lambda amount: assets_rate([amount/2, 0], state), deposits
                )), 
                color='red', 
                label='1', 
                alpha=0.8
               )
ax.plot(list(map(lambda d: d + 57, deposits)), 
                list(map(
                    lambda amount: assets_rate([0, amount/2], state), deposits
                )), 
                color='blue', 
                label='2', 
                alpha=0.8
               )
ax.legend(loc='upper left')
ax.set_title('Deposit')
ax.set_ylabel('rate')
ax.set_xlabel('steps')
ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.5, color='black')
plt.show()
```

```python
import pool
import importlib
from copy import deepcopy
from functools import reduce
importlib.reload(pool)

weights = [0.5, 0.5]
state = pool.create(weights=weights, max_allocation=0.8, weight_amplitude=0.65, slippage_rate=2, fee_max=10000000, fee_base=0)
pool.proportional_deposit(200, state)

total_lp_1 = reduce(lambda acc, amount: pool.deposit([amount, 0], state) + acc, range(1, 40, 1))
total_lp_2 = reduce(lambda acc, amount: pool.deposit([0, amount], state) + acc, range(1, 40, 1))
lp_diff = (total_lp_1 + total_lp_2) - sum(range(1, 40, 1)) * 2
assert (0.001 > lp_diff > 0), f"lp diff:{lp_diff}"
print(lp_diff)
```

```python
import pool
import importlib
from copy import deepcopy
import matplotlib.pyplot as plt
importlib.reload(pool)

weights = [0.5, 0.5]
state = pool.create(weights=weights, max_allocation=0.8, weight_amplitude=0.75, slippage_rate=1, fee_max=0.03, fee_base=0.1)
pool.proportional_deposit(1000, state)

%matplotlib widget
deposits = list(range(1, 4000, 1))

def deposit_rate(amounts: [int], state: pool.PoolState):
    a = pool.deposit(amounts, state)
    return state.assets_total_supply / state.total_supply

def withdraw_rate(amounts: [int], state: pool.PoolState):
    a = pool.withdraw(amounts, state)
    return state.assets_total_supply / state.total_supply

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(deposits, 
                list(map(
                    lambda amount: deposit_rate([0, 1], state), deposits
                )),
                color='red', 
                label='deposit', 
                alpha=0.8
               )
print(state)
ax.plot(list(map(lambda i: i + 4000, deposits)), 
                list(map(
                    lambda amount: withdraw_rate([0, 1], state), deposits
                )),
                color='blue', 
                label='withdraw', 
                alpha=0.8
               )
print(state)
ax.legend(loc='upper left')
ax.set_title('LP rate')
ax.set_ylabel('rate')
ax.set_xlabel('steps')
ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.5, color='black')
plt.show()
```
